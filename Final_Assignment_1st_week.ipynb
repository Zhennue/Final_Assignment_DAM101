```python
import os
from youtubesearchpython import VideosSearch
import yt_dlp

# Step 1: Search YouTube
query = "top trending English songs"
max_results = 50

videosSearch = VideosSearch(query, limit=max_results)
search_results = videosSearch.result()["result"]

# Step 2: Prepare output path
output_dir = '/content/drive/MyDrive/SS2025/DAM101/Assignment_4/Data_set/English'
os.makedirs(output_dir, exist_ok=True)

# Step 3: yt-dlp options
ydl_opts = {
    'format': 'bestaudio/best',
    'outtmpl': os.path.join(output_dir, '%(title)s.%(ext)s'),
    'cookiefile': 'cookies.txt',
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'mp3',
        'preferredquality': '192', # bitrate
    }],
    'quiet': False, # show messages while downloading
    'noplaylist': True, # specific video is downloaded, not an entire series of videos.
}

# Step 4: Download each video if not already downloaded
with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    for result in search_results:
        url = result["link"]
        title = result["title"]
        filename = os.path.join(output_dir, f"{title}.mp3")

        if os.path.exists(filename):
            print(f"Already downloaded: {title}")
            continue

        try:
            ydl.download([url])
        except Exception as e:
            print(f"Error downloading {url}: {e}")
```

```python
import os
import subprocess
from pathlib import Path

directory = "/content/drive/MyDrive/SS2025/DAM101/Assignment_4/Data_set/Dzongkha"

def is_valid_mp3(file_path):
    try:
        result = subprocess.run(
            ['ffmpeg', '-v', 'error', '-i', file_path, '-f', 'null', '-'],
            stdout=subprocess.PIPE, stderr=subprocess.PIPE
        )
        return result.returncode == 0
    except Exception:
        return False

# Go through all mp3s and rename broken ones
for file in os.listdir(directory):
    if file.lower().endswith(".mp3"):
        path = os.path.join(directory, file)
        print(f"[CHECKING] {file}")
        if not is_valid_mp3(path):
            broken_name = os.path.splitext(file)[0] + "_broken.mp3"
            broken_path = os.path.join(directory, broken_name)
            os.rename(path, broken_path)
            print(f"[RENAMED] {file} -> {broken_name}")
```

```python
import os
import subprocess
import tempfile
import gc # Garbage collector
from pydub import AudioSegment
from pydub.silence import detect_nonsilent

os.environ["TF_FORCE_GPU_ALLOW_GROWTH"] = "true"

# === CONFIG ===
CLIP_DURATION_MS = 10 * 1000  # 10 seconds
SILENCE_THRESH_DB = -40
MIN_SILENCE_LEN_MS = 1000 # 1 second break
SUPPORTED_EXT = ('.mp3', '.wav', '.m4a')
DURATION_THRESHOLD_SEC = 300  # 5 min
SEGMENT_COUNT_THRESHOLD = 3
MIN_SEGMENT_MS = 10000  # 10s

# === Spleeter Separation ===
def spleeter_vocals_once(input_path, tmp_output_dir):
    base = os.path.splitext(os.path.basename(input_path))[0]
    output_dir = os.path.join(tmp_output_dir, base)
    vocals_path = os.path.join(output_dir, "vocals.wav")

    if os.path.exists(vocals_path):
        return vocals_path

    cmd = ["spleeter", "separate", "-p", "spleeter:2stems", "-o", tmp_output_dir, "--mwf", input_path]
    try:
        print(f"[SPLEETER] Running: {input_path}")
        subprocess.run(cmd, check=True, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
        return vocals_path if os.path.exists(vocals_path) else None
    except Exception as e:
        print(f"[ERROR] Spleeter failed on {base}: {e}")
        return None

# === Clip Extractor ===
def extract_10s_clips(vocal_audio, base, output_root):
    segments = detect_nonsilent(vocal_audio, min_silence_len=MIN_SILENCE_LEN_MS, silence_thresh=SILENCE_THRESH_DB)
    segments = [(start, end) for start, end in segments if end - start >= 5000]

    clip_index = 0
    for i, (start, end) in enumerate(segments):
        segment = vocal_audio[start:end]
        while len(segment) >= CLIP_DURATION_MS:
            clip = segment[:CLIP_DURATION_MS]
            output_path = os.path.join(output_root, f"{base}_clip{clip_index+1}_10s.wav")
            clip.export(output_path, format="wav")
            print(f"[OK] Saved clip: {output_path}")
            clip_index += 1
            segment = segment[CLIP_DURATION_MS:]

# === Auto Detection & Split ===
def process_audio_auto(audio_path, output_root, tmp_output_dir):
    base = os.path.splitext(os.path.basename(audio_path))[0]
    print(f"[PROCESSING] {base}")

    try:
        audio = AudioSegment.from_file(audio_path)
    except Exception as e:
        print(f"[ERROR] Could not load file: {e}")
        return

    duration_sec = len(audio) / 1000
    segments = detect_nonsilent(audio, min_silence_len=2000, silence_thresh=SILENCE_THRESH_DB)
    long_segments = [(start, end) for start, end in segments if end - start >= MIN_SEGMENT_MS]

    is_album = duration_sec > DURATION_THRESHOLD_SEC or len(long_segments) >= SEGMENT_COUNT_THRESHOLD
    print(f"[DETECT] {base} classified as {'Album' if is_album else 'Song'}")

    if is_album:
        for i, (start, end) in enumerate(long_segments):
            try:
                with tempfile.NamedTemporaryFile(suffix=".wav", delete=False) as tmp_seg:
                    audio[start:end].export(tmp_seg.name, format="wav")
                    vocal_path = spleeter_vocals_once(tmp_seg.name, tmp_output_dir)
                os.unlink(tmp_seg.name)

                if vocal_path and os.path.exists(vocal_path):
                    vocal_audio = AudioSegment.from_file(vocal_path)
                    extract_10s_clips(vocal_audio, f"{base}_song{i+1}", output_root)
                    gc.collect()
            except Exception as e:
                print(f"[ERROR] Album segment failed: {e}")
    else:
        try:
            vocal_path = spleeter_vocals_once(audio_path, tmp_output_dir)
            if not vocal_path:
                return
            vocal_audio = AudioSegment.from_file(vocal_path)
            extract_10s_clips(vocal_audio, base, output_root)
            gc.collect()
        except Exception as e:
            print(f"[ERROR] Song processing failed: {e}")

# === Directory Processor ===
def process_all_files_one_by_one(dataset_dir, output_dir):
    os.makedirs(output_dir, exist_ok=True)

    for file in sorted(os.listdir(dataset_dir)):
        if not file.lower().endswith(SUPPORTED_EXT):
            continue

        full_path = os.path.join(dataset_dir, file)
        base = os.path.splitext(file)[0]

        already_processed = any(
            f.startswith(base) and f.endswith("_10s.wav")
            for f in os.listdir(output_dir)
        )
        if already_processed:
            print(f"[SKIP] {file} — already processed.")
            continue

        try:
            with tempfile.TemporaryDirectory() as tmp_output_dir:
                process_audio_auto(full_path, output_dir, tmp_output_dir)
        except Exception as e:
            print(f"[FATAL ERROR] Could not process {file}: {e}")

# === INPUT PATHS ===
input_root = "/content/drive/MyDrive/SS2025/DAM101/Assignment_4/Data_set/Dzongkha/"
output_root = "/content/drive/MyDrive/SS2025/DAM101/Assignment_4/Processed_Dataset/Dzongkha/"

# === RUN ===
process_all_files_one_by_one(input_root, output_root)
```

```python
import os
import torch
import torchaudio
import soundfile as sf
import librosa
import numpy as np
import glob
import matplotlib.pyplot as plt

TARGET_LENGTH = 10  # seconds
SAMPLE_RATE = 16000
TARGET_SAMPLES = SAMPLE_RATE * TARGET_LENGTH
MIN_SONG_GAP = 3.0
MIN_SEGMENT_DURATION = 30.0

def extract_10s_clip(waveform, sr):
    if waveform.shape[0] > 1:
        waveform = waveform.mean(dim=0, keepdim=True)
    if sr != SAMPLE_RATE:
        waveform = torchaudio.functional.resample(waveform, sr, SAMPLE_RATE)
    waveform = waveform.squeeze()
    if waveform.shape[0] < TARGET_SAMPLES:
        pad = torch.zeros(TARGET_SAMPLES - waveform.shape[0])
        waveform = torch.cat([waveform, pad])
    else:
        waveform = waveform[:TARGET_SAMPLES]
    return waveform.numpy()

def extract_audio_features(y, sr, output_prefix):
    # Mel Spectrogram
    mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
    mel_db = librosa.power_to_db(mel, ref=np.max)
    np.save(output_prefix + "_mel.npy", mel_db)

    # MFCC
    mfcc = librosa.feature.mfcc(S=mel_db, sr=sr, n_mfcc=13)
    np.save(output_prefix + "_mfcc.npy", mfcc)

    # Chroma
    chroma = librosa.feature.chroma_stft(y=y, sr=sr)
    np.save(output_prefix + "_chroma.npy", chroma)

    # Spectral Contrast
    contrast = librosa.feature.spectral_contrast(y=y, sr=sr)
    np.save(output_prefix + "_contrast.npy", contrast)

    print(f"[OK] Features saved for: {output_prefix}")

def detect_song_segments(y, sr):
    energy = librosa.feature.rms(y=y)[0]
    frames = range(len(energy))
    times = librosa.frames_to_time(frames, sr=sr)

    silence_threshold = 0.01
    silent_frames = energy < silence_threshold
    segments = []
    start = 0
    for i in range(1, len(silent_frames)):
        if silent_frames[i] and not silent_frames[i - 1]:
            end = times[i]
            if end - start >= MIN_SEGMENT_DURATION:
                segments.append((start, end))
            start = times[i]
    if times[-1] - start >= MIN_SEGMENT_DURATION:
        segments.append((start, times[-1]))
    return segments if segments else [(0, min(600, librosa.get_duration(y=y, sr=sr)))]

def process_audio_file(file_path, output_dir, index_prefix=""):
    try:
        # Skip already processed files
        existing_outputs = glob.glob(os.path.join(output_dir, f"{index_prefix}_*.wav"))
        if existing_outputs:
            print(f"[SKIP] Already processed: {file_path}")
            return

        print(f"[INFO] Processing {file_path}")
        y, sr = librosa.load(file_path, sr=None, mono=True)
        duration = librosa.get_duration(y=y, sr=sr)

        if duration < 60:
            waveform, sr_loaded = torchaudio.load(file_path)
            clip = extract_10s_clip(waveform, sr_loaded)
            output_prefix = os.path.join(output_dir, f"{index_prefix}_01")
            sf.write(output_prefix + ".wav", clip, SAMPLE_RATE)
            extract_audio_features(clip, SAMPLE_RATE, output_prefix)
        else:
            segments = detect_song_segments(y, sr)
            for idx, (start, end) in enumerate(segments):
                segment = y[int(start * sr):int((start + TARGET_LENGTH) * sr)]
                if len(segment) < 1000:
                    continue
                output_prefix = os.path.join(output_dir, f"{index_prefix}_{idx+1:02d}")
                sf.write(output_prefix + ".wav", segment, sr)
                extract_audio_features(segment, sr, output_prefix)

    except Exception as e:
        print(f"[ERROR] {file_path}: {e}")

# === Main Loop ===
input_root = "/content/drive/MyDrive/SS2025/DAM101/Assignment_4/Processed_Dataset/Dzongkha"
output_root = "/content/drive/MyDrive/SS2025/DAM101/Assignment_4/Melspectogram/Dzongkha"
os.makedirs(output_root, exist_ok=True)

for album_dir, _, files in os.walk(input_root):
    audio_files = [f for f in files if f.endswith((".mp3", ".wav", ".mp4"))]
    if len(audio_files) < 2:
        print(f"[SKIP] Album '{album_dir}' has fewer than 2 songs.")
        continue

    relative_album_path = os.path.relpath(album_dir, input_root)
    album_output_dir = os.path.join(output_root, relative_album_path)
    os.makedirs(album_output_dir, exist_ok=True)

    for file in audio_files:
        full_path = os.path.join(album_dir, file)
        index_prefix = os.path.splitext(file)[0]
        process_audio_file(full_path, album_output_dir, index_prefix=index_prefix)
```
